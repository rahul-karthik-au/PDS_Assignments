```python
import numpy as np
import pandas as pd

student_data=pd.read_csv('StudentsPerformance.csv')
student_data
```

```python
student_data.drop('lunch', axis=1, inplace=True)

csv_filename = 'clean_student_data.csv'
df = pd.DataFrame(student_data)
df.to_csv(csv_filename, index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

race_counts = student_data['race/ethnicity'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Race/Ethnicity')
plt.axis('equal')
plt.show()
```

```python
parental_level_of_education_count = student_data['parental level of education'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(parental_level_of_education_count, labels=parental_level_of_education_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Parental Level Of Education')
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns

avg_math_scores = student_data.groupby(['race/ethnicity', 'gender'])['math score'].mean().unstack()

plt.figure(figsize=(10, 6))
avg_math_scores.plot(kind='bar',color=sns.palettes.mpl_palette('Dark2'),  alpha=0.7)
plt.title('Average Math Scores by Gender and Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Math Score')

plt.legend(title='Gender')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
avg_reading_scores = student_data.groupby(['race/ethnicity', 'gender'])['reading score'].mean().unstack()

plt.figure(figsize=(10, 6))

avg_reading_scores.plot(kind='bar', color=['blue', 'orange'], alpha=0.7)

plt.title('Average Reading Scores by Gender and Race/Ethnicity')

plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Reading Score')

plt.legend(title='Gender')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
avg_writing_scores = student_data.groupby(['race/ethnicity', 'gender'])['writing score'].mean().unstack()

plt.figure(figsize=(10, 6))

avg_reading_scores.plot(kind='bar', alpha=0.7)

plt.title('Average Reading Scores by Gender and Race/Ethnicity')

plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Reading Score')

plt.legend(title='Gender')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
above_80_math_count = student_data[student_data['math score'] > 80].groupby(['test preparation course','race/ethnicity'])['math score'].size().unstack()

plt.figure(figsize=(20, 6))
above_80_math_count.plot(kind='bar')
plt.title('Count of Math Scores Above 80 by Test Preparation Course and Race/Ethincity')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
