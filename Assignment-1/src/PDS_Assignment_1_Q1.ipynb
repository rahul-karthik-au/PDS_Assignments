```python
import numpy as np
import pandas as pd
```

```python
frailty_data=pd.read_csv('frailty_data.csv')
frailty_data
```

```python
frailty_data['Frailty'] = frailty_data['Frailty'].replace('Y',1)
```

```python
frailty_data['Frailty'] = frailty_data['Frailty'].replace('N',0)
```

```python
frailty_data['Frailty'] = frailty_data['Frailty'].astype(int)
```

```python
frailty_data
```

```python
csv_filename = 'clean_frailty_data.csv'
df = pd.DataFrame(frailty_data)
df.to_csv(csv_filename, index=False)
```
