```python
import numpy as np
import pandas as pd

data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train.csv')
dup_data=data
dup_data
```

Removed the Unnamed Column

```python
data.drop('Unnamed: 0',axis = 1,inplace = True)
```

Removed the units from the attributes and kept only the numerical values.

For This first I have converted datatype to string and performed split operation to separate numerical values.
After seperation changed the data type as Float or Integer according to the need.

Mileage - Float (since for ex: 19.67, 13, 20.77, ...)

Engine - Integer (since for ex: 1582, 1199, 1248, ...)

Power - Float (since for ex: 126.2, 88.7, 88.76, ...)

New_Price - Float (since for ex: 8.61, 7.88, ...)






```python
data['Mileage'] = data['Mileage'].astype(str)
data['Mileage'] = data['Mileage'].str.split().str[0].astype(float)
data['Engine'] = data['Engine'].astype(str)
data['Engine'] = data['Engine'].str.split().str[0].astype(float)
data['Engine'] = data['Engine'].fillna(0).astype(int)
data['Power'] = data['Power'].astype(str)
data['Power'] = data['Power'].str.split().str[0].astype(float)

data['New_Price'] = data['New_Price'].astype(str)
data['New_Price'] = data['New_Price'].str.split().str[0].astype(float)

data.dtypes
data
```

Filled the missing value of Seats by the ***mode*** of the Cars by Name.

Removed the other rows where we could not find single row with similar Name.

```python
data['Seats'] = data.groupby('Name')['Seats'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x)
data = data.dropna(subset=['Seats'])
data.loc[:, 'Seats'] = data['Seats'].astype(int)
data
```

Found the distinct values of Fuel Type and Transmission so that it will help us in changing it to a Categorical Value.

```python
distinct_FuelType = data['Fuel_Type'].unique()
distinct_FuelType
```

```python
distinct_Transmission=data['Transmission'].unique()
distinct_Transmission
```

Fuel Type Categorical Value changed to Numerical Value:

Since we have 3 distinct vaules in Fuel Type i.e.,

Diesel:0

Petrol:1

Electric:2

```python
Fuel_Type_Mapping = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
data.loc[:, 'Fuel_Type'].replace(Fuel_Type_Mapping, inplace=True)
data
```

Transmission Categorical Value changed to Numerical Value:

Since we have 2 distinct vaules in Transmission Type i.e.,

Manuel:0

Automatic:1

```python
Transmission_Mapping = {'Manual': 0, 'Automatic': 1}
data.loc[:, 'Transmission'].replace(Transmission_Mapping, inplace=True)
data
```

Finding the number of NaN and zero value of each column.

```python
nan_zero_count_Year = data['Year'].isna().sum() + (data['Year'] == 0).sum()
nan_zero_count_Kilometers_Driven = data['Kilometers_Driven'].isna().sum() + (data['Kilometers_Driven'] == 0).sum()
nan_zero_count_Fuel_Type = data['Fuel_Type'].isna().sum()
nan_zero_count_Transmission = data['Transmission'].isna().sum()
nan_zero_count_Mileage = data['Mileage'].isna().sum() + (data['Mileage'] == 0).sum()
nan_zero_count_Engine = data['Engine'].isna().sum() + (data['Engine'] == 0).sum()
nan_zero_count_Power = data['Power'].isna().sum() + (data['Power'] == 0).sum()
nan_zero_count_Seats = data['Seats'].isna().sum() + (data['Seats'] == 0).sum()
nan_zero_count_New_Price = data['New_Price'].isna().sum() + (data['New_Price'] == 0).sum()
nan_zero_count_Price = data['Price'].isna().sum() + (data['Price'] == 0).sum()


print("Nan or Zero Count of Year:",nan_zero_count_Year)
print("Nan or Zero Count of Kilometers Driven:",nan_zero_count_Kilometers_Driven)
print("Nan or Zero Count of Fuel Type:",nan_zero_count_Fuel_Type)
print("Nan or Zero Count of Transmissiom:",nan_zero_count_Transmission)
print("Nan or Zero Count of Mileage:",nan_zero_count_Mileage)
print("Nan or Zero Count of Engine:",nan_zero_count_Engine)
print("Nan or Zero Count of Power:",nan_zero_count_Power)
print("Nan or Zero Count of Seats:",nan_zero_count_Seats)
print("Nan or Zero Count of New Price:",nan_zero_count_New_Price)
print("Nan or Zero Count of Price:",nan_zero_count_Price)
```

Finding the mean Mileage value which is group by Owner Type ,Fuel Type and Transmission.

I think these are the factors which Mileage is dependent.

```python
mean_mileage = data.groupby(['Owner_Type', 'Fuel_Type', 'Transmission'])['Mileage'].mean()
mean_mileage
```

Filling with the mean value by grouping into a particular category by Owner Type ,Fuel Type and Transmission.

```python
data['Mileage'] = data.groupby(['Owner_Type', 'Fuel_Type', 'Transmission'])['Mileage'].transform(lambda x: x.fillna(x.mean()).replace(0))
```

```python
data
```

There are still NaN value because Fuel Type 'Electric' did not have Mileage.

```python
nan_zero_count_Mileage = data['Mileage'].isna().sum() + (data['Mileage'] == 0).sum()
print("Nan or Zero Count of Mileage:",nan_zero_count_Mileage)
```

After filling that value to 10 there is no zero or Nan Values.

```python
data['Mileage']=data['Mileage'].fillna(10)
```

```python
nan_zero_count_Mileage = data['Mileage'].isna().sum() + (data['Mileage'] == 0).sum()
print("Nan or Zero Count of Mileage:",nan_zero_count_Mileage)
```

Filled Empty Values of Engine with value of Engine whose Name are similar.

```python
name_to_engine = data[data['Engine'] != 0].groupby('Name')['Engine'].first().to_dict()
data['Engine'] = data.groupby('Name')['Engine'].transform(lambda x: x if x.iloc[0] != 0 else name_to_engine.get(x.name, 0))
data = data[data['Engine'] != 0]
```

Filled Empty Power Value rows with value of Power with help of mode of Power value with similar Name

```python
name_to_power = data[data['Power'] != 0].groupby('Name')['Power'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None).to_dict()
data['Power'] = data.groupby('Name')['Power'].transform(lambda x: x.replace(0).fillna(name_to_power.get(x.name)))
```

Replaced New_Price with Price Value:

```python
data['New_Price']=data['New_Price'].fillna(data['Price'])
```

Create new Age Column.

```python
current_year = 2024
clean_data['Age'] = current_year - clean_data['Year']
clean_data
```

Created new column called Avg.KMs Per Year withe the help of age calculated above.

```python
clean_data['Avg.KMs Per Year']=clean_data['Kilometers_Driven']/clean_data['Age']
clean_data
```

Remaned the Avg.KMs Per Year to Kilometers_Driven Per Year

```python
clean_data=clean_data.rename(columns={'Avg.KMs Per Year':'Kilometers_Driven Per Year'})
```

Created a new column called Fuel Efficiency with the help of Power and Engine.

```python
clean_data['Fuel_Efficiency'] = clean_data['Power'] / clean_data['Engine']
clean_data
```

Summarize Grouped By Location.
